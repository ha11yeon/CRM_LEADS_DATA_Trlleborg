```python
# üì¶ ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)

# üìÇ Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv('../data/leads_df_cleaned.csv')

# üéØ CRM ÌïµÏã¨ Ïó¥ ÏÑ†ÌÉù
cols = ['Lead_ID', 'Name', 'Account_Party_Name', 'Main_Contact_Person_Name',
        'Contact_Information_Gender', 'Account_Status']
df_crm = df[cols]

# ‚ùì Í≤∞Ï∏°Ïπò ÌôïÏù∏
print("üîç Í≤∞Ï∏°Ïπò:\n", df_crm.isnull().sum())


```

```python
# üìä Î¶¨Îìú Ïàò Î∞è Í≥†Í∞ùÏÇ¨ ÏÉÅÌÉú Î∂ÑÌè¨
print("Í≥†Ïú† Î¶¨Îìú Ïàò:", df_crm['Lead_ID'].nunique())
print("\nÍ≥†Í∞ùÏÇ¨ ÏÉÅÌÉú Î∂ÑÌè¨:\n", df_crm['Account_Status'].value_counts())

```

```python

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Ïòà: ÎßëÏùÄ Í≥†Îîï)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
# üìà Í≥†Í∞ùÏÇ¨ ÏÉÅÌÉú ÏãúÍ∞ÅÌôî
sns.countplot(data=df_crm, x='Account_Status')
plt.title('Í≥†Í∞ùÏÇ¨ ÏÉÅÌÉú Î∂ÑÌè¨')
plt.xlabel('ÏÉÅÌÉú')
plt.ylabel('Î¶¨Îìú Ïàò')
plt.xticks(rotation=45)
plt.show()
```

```python

# üìà Îã¥ÎãπÏûê ÏÑ±Î≥Ñ Î∂ÑÌè¨
sns.countplot(data=df_crm, x='Contact_Information_Gender')
plt.title('Îã¥ÎãπÏûê ÏÑ±Î≥Ñ Î∂ÑÌè¨')
plt.xlabel('ÏÑ±Î≥Ñ ÏΩîÎìú')
plt.ylabel('Î¶¨Îìú Ïàò')
plt.show()
```
