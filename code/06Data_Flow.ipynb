```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
# Îç∞Ïù¥ÌÑ∞ Î°úÎìú
df = pd.read_csv('../data/leads.csv')

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Ïòà: ÎßëÏùÄ Í≥†Îîï)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
```

```python
# ÏòÅÏóÖ Í∏∞Ìöå Î∞è ÏòàÏ∏° Ï†ïÎ≥¥ Í¥ÄÎ†® Ïó¥ Ï∂îÏ∂ú
sales_cols = ['Priority_KUT', 'Expected_Revenue_Value', 'Sealing_Demand_Amount__Currency', 'Score', 'Contact_Information_Engagement_Score', 'Survey_Score']
df_sales = df[sales_cols]
df_sales.head()
```

```python
df_sales.isnull().sum()
```

## üìà Î¶¨Îìú Ïö∞ÏÑ†ÏàúÏúÑ(Priority_KUT) Î∂ÑÌè¨

```python
sns.histplot(df_sales['Priority_KUT'].dropna(), bins=20, kde=True)
plt.title('Î¶¨Îìú Ïö∞ÏÑ†ÏàúÏúÑ Î∂ÑÌè¨')
plt.xlabel('Priority_KUT')
plt.ylabel('Count')
plt.show()
```

# Ïö∞ÏÑ†ÏàúÏúÑÎßå ÏùòÎØ∏Í∞Ä ÏûàÏùåÏùÑ ÌôïÏù∏

## üí∞ ÏòàÏÉÅ Îß§Ï∂ú Î∂ÑÌè¨(Expected_Revenue_Value)

```python
sns.histplot(df_sales['Expected_Revenue_Value'].dropna(), bins=30, kde=True)
plt.title('ÏòàÏÉÅ Îß§Ï∂ú Î∂ÑÌè¨')
plt.xlabel('Expected Revenue')
plt.ylabel('Î¶¨Îìú Ïàò')
plt.show()
```

## ‚öôÔ∏è CRM ÌôúÎèôÏÑ± Ï†êÏàò (Engagement Score)

```python
sns.histplot(df_sales['Contact_Information_Engagement_Score'].dropna(), bins=20, kde=True)
plt.title('Engagement Score Î∂ÑÌè¨')
plt.xlabel('Score')
plt.ylabel('Î¶¨Îìú Ïàò')
plt.show()
```

## üìù Í≥†Í∞ù ÎßåÏ°±ÎèÑ Ï†êÏàò (Survey Score)

```python
sns.histplot(df_sales['Survey_Score'].dropna(), bins=20, kde=True)
plt.title('Í≥†Í∞ù ÎßåÏ°±ÎèÑ Ï†êÏàò Î∂ÑÌè¨')
plt.xlabel('Survey Score')
plt.ylabel('Î¶¨Îìú Ïàò')
plt.show()
```
