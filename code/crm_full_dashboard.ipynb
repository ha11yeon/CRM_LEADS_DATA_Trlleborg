```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
%matplotlib inline
```

## üìÇ Îç∞Ïù¥ÌÑ∞ Î°úÎìú

```python
df = pd.read_csv('../data/leads_df_cleaned.csv')
df.shape
```

## üë• Í≥†Í∞ù ÏãùÎ≥Ñ Î∞è ÌîÑÎ°úÌïÑ Î∂ÑÏÑù

```python

# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Ïòà: ÎßëÏùÄ Í≥†Îîï)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
```

```python
customer_cols = ['Lead_ID', 'Name', 'Account_Party_Name', 'Main_Contact_Person_Name',
                 'Contact_Information_Gender', 'Account_Status']
df_customer = df[customer_cols].dropna()
sns.countplot(data=df_customer, x='Account_Status')
plt.title('Í≥†Í∞ùÏÇ¨ ÏÉÅÌÉú Î∂ÑÌè¨')
plt.show()
```

## üîÅ Î¶¨Îìú ÏÉÅÌÉú Î∞è Îã®Í≥Ñ Î∂ÑÏÑù

```python
status_cols = ['Status_Text', 'Lead_Life_Cycle_Status_Code', 'Main_Contact_Person_Status']
df_status = df[status_cols].dropna()
sns.countplot(data=df_status, x='Status_Text')
plt.title('Î¶¨Îìú ÏÉÅÌÉú Î∂ÑÌè¨')
plt.xticks(rotation=45)
plt.show()
```

## üì£ ÎßàÏºÄÌåÖ/ÏÑ∏ÏùºÏ¶à Ï±ÑÎÑê Î∂ÑÏÑù

```python
mkt_cols = ['ContactSource', 'Owner_Party_Name']
df_mkt = df[mkt_cols].dropna()
sns.countplot(data=df_mkt, x='ContactSource', order=df_mkt['ContactSource'].value_counts().index)
plt.title('Î¶¨Îìú Ïú†ÏûÖ Ï±ÑÎÑê')
plt.xticks(rotation=45)
plt.show()
```
