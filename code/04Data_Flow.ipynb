```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
%matplotlib inline
```

```python
# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv('../data/leads_df_cleaned.csv')

# Ï£ºÏöî ÏÉÅÌÉú Í¥ÄÎ†® Ïó¥Îßå Ï∂îÏ∂ú
status_cols = ['Status_Text',  'Lead_Life_Cycle_Status_Code', 'Main_Contact_Person_Status']
df_status = df[status_cols]
df_status.head()
```

```python
df_status.isnull().sum()


# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (Ïòà: ÎßëÏùÄ Í≥†Îîï)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ
```

## üìà Î¶¨Îìú ÏÉÅÌÉúÎ≥Ñ Î∂ÑÌè¨

```python
sns.countplot(data=df_status, x='Status_Text')
plt.title('Î¶¨Îìú ÏÉÅÌÉú Î∂ÑÌè¨')
plt.xlabel('Status_Text')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## üìà Lead_Life_Cycle_Status_Code Î∂ÑÌè¨
